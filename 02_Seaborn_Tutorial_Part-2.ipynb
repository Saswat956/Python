```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
df.head(5)
```

## FacetGrid

```python
fg = sns.FacetGrid(df,row="time",col="day")
fg.map(sns.scatterplot,"tip","total_bill")
```

```python
fg = sns.FacetGrid(df,row="time",col="day",hue="smoker")
fg.map(sns.regplot,"tip","total_bill")
```

```python
fg = sns.FacetGrid(df,row="time",col="day")
fg.map(sns.violinplot,"tip","total_bill")
```

```python
df.head(2)
```

```python
fg = sns.FacetGrid(df,row="smoker",col="day")
fg.map(sns.kdeplot,"tip","total_bill")
```

## PairGrid

```python
df.head(1)
```

```python
pg = sns.PairGrid(df,hue="smoker")
pg.map_upper(sns.violinplot)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.regplot)
```

```python
pg = sns.PairGrid(df,hue="smoker")
pg.map_upper(sns.violinplot)
pg.map_diag(sns.barplot)
pg.map_lower(sns.kdeplot)
```

### KdePlot 1D (univariate) Vs 2D (Bivariate)

```python
df.head(1)
```

```python
sns.kdeplot(df.total_bill)
```

```python
sns.kdeplot(df.total_bill,df.tip)
```

```python
sns.kdeplot(df.total_bill,df.tip,shade=True)
```

```python
sns.kdeplot(df.total_bill,df.tip,shade=True,cbar=True)
```

```python
sns.kdeplot(df.total_bill,df.tip,shade=True,cbar=True)
sns.rugplot(df.total_bill)
```

```python
sns.distplot(df.total_bill,rug=True)
```

### RugPlot

```python
sns.rugplot(df.total_bill)
```

```python
sns.distplot(df.total_bill)
sns.rugplot(df.total_bill)
```

```python
sns.distplot(df.total_bill,kde=False)
sns.rugplot(df.total_bill,axis="y")
```
